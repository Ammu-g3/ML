```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv(r"C:\Users\LENOVO\Downloads\homeprices.csv")
```

```python
data
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(data.area,data.price,color='blue',marker='*')
```

```python
x=data.drop('price',axis=1) # feature
y=data['price'] #target
```

```python
x
```

```python
y
```

```python
model=LinearRegression() #creating object of the Linear Regression class
model.fit(x,y) #model tarining
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x)
#3000 sqft area enkil and y=10 then y_pred=m*3000+c
```

```python
df1=pd.DataFrame({'True y values':y, 'predicted y values':y_pred})
pd.concat([x,df1],axis=1)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.xlabel('Area')
plt.ylabel('Price')
plt.plot(x,df1['predicted y values'], color='red')
```
